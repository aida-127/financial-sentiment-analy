```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import talib  # ‚úÖ This will work now!
import yfinance as yf
from datetime import datetime, timedelta

print("‚úÖ TA-Lib imported successfully!")
```

```python
# Cell 2: Download stock data with better error handling
print("üì• Downloading stock data...")

ticker = "AAPL"
start_date = "2020-01-01"
end_date = "2024-01-01"

try:
    # Download with timeout handling
    stock_data = yf.download(
        ticker, 
        start=start_date, 
        end=end_date, 
        progress=False,
        timeout=30  # Increase timeout
    )
    
    if stock_data.empty:
        print("‚ùå Download failed - creating sample data for demonstration")
        # Create sample data so you can continue
        dates = pd.date_range(start=start_date, end=end_date, freq='D')
        stock_data = pd.DataFrame({
            'Open': np.random.uniform(100, 200, len(dates)),
            'High': np.random.uniform(150, 250, len(dates)),
            'Low': np.random.uniform(80, 150, len(dates)),
            'Close': np.random.uniform(120, 180, len(dates)),
            'Adj Close': np.random.uniform(120, 180, len(dates)),
            'Volume': np.random.randint(1000000, 50000000, len(dates))
        }, index=dates)
        print("‚úÖ Created sample stock data for demonstration")
    else:
        print(f"‚úÖ Downloaded {ticker} data: {stock_data.shape}")
        
except Exception as e:
    print(f"‚ùå Download error: {e}")
    print("Creating sample data to continue...")
    # Create sample data
    dates = pd.date_range(start=start_date, end=end_date, freq='D')
    stock_data = pd.DataFrame({
        'Open': np.random.uniform(100, 200, len(dates)),
        'High': np.random.uniform(150, 250, len(dates)),
        'Low': np.random.uniform(80, 150, len(dates)),
        'Close': np.random.uniform(120, 180, len(dates)),
        'Adj Close': np.random.uniform(120, 180, len(dates)),
        'Volume': np.random.randint(1000000, 50000000, len(dates))
    }, index=dates)

print(f"Final data shape: {stock_data.shape}")
print(stock_data.head())
```

```python
# Cell 3: Calculate Technical Indicators (FIXED)
print("üîÑ Calculating technical indicators...")

# Fix: Convert from (1006, 1) to (1006,) - 1D array
close_prices = stock_data['Close'].values.flatten()  # ‚Üê ADD .flatten()
# OR
close_prices = stock_data['Close'].iloc[:, 0].values  # ‚Üê Get first column values

print(f"‚úÖ Close prices shape: {close_prices.shape}")

# Now calculate indicators
stock_data['SMA_20'] = talib.SMA(close_prices, timeperiod=20)
stock_data['SMA_50'] = talib.SMA(close_prices, timeperiod=50)
stock_data['RSI'] = talib.RSI(close_prices, timeperiod=14)
macd, macd_signal, macd_hist = talib.MACD(close_prices)
stock_data['MACD'] = macd
stock_data['MACD_Signal'] = macd_signal
stock_data['MACD_Hist'] = macd_hist

print("‚úÖ TA-Lib indicators calculated successfully!")
print(stock_data[['Close', 'SMA_20', 'SMA_50', 'RSI', 'MACD']].tail())
```
